# Stima di $\pi$ con il metodo MonteCarlo

```python
N=[10,50,100,500,1000,5000,10000,50000,100000] #Valori di N per cui genero i punti
pi=[0.]*len(N) #Valore di pi per ogni N

for i in range(len(N)):
  x=[0.]*N[i]
  y=[0.]*N[i]
  k=0
  for j in range(N[i]): #Genero i punti
    x[j]=random.uniform(-1.0,1.0)
    y[j]=random.uniform(-1.0,1.0)
    if x[j]*x[j]+y[j]*y[j]<=1: #Controllo che il punto sia nel cerchio
      k=k+1 #Conto i punti nel cerchio
  pi[i]=4*k/N[i] #Aggiorno il valore di pi per ogni N
```

```python
x_c=[0.]*k
y_c=[0.]*k
x_s=[0.]*(N[len(N)-1]-k)
y_s=[0.]*(N[len(N)-1]-k)
j=0
h=0

for i in range(N[len(N)-1]):
  if x[i]*x[i]+y[i]*y[i]<=1: #Separo i punti nel cerchio...
    x_c[j]=x[i]
    y_c[j]=y[i]
    j=j+1 
  else: #...da quelli fuori
    x_s[h]=x[i]
    y_s[h]=y[i]
    h=h+1
```
