```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
fnameAn = 'dataAn.txt'

vxAn,vyAn, tAn = np.loadtxt(fnameAn, unpack=True, usecols=(0,1,2))

fnameRK = 'dataRK.txt'

vxRK,vyRK, tRK = np.loadtxt(fnameRK, unpack=True, usecols=(0,1,2))
```

```python
fig1 = plt.figure(figsize=(8, 6), dpi=400)

plt.plot(tAn, vxAn, label="$v_x$ Analytical", ls = "-")
plt.plot(tRK, vxRK, label="$v_x$ RungeKutta", ls = "-.")
plt.xlabel("Time [s]")
plt.ylabel("Velocity [m/s]")
plt.legend()
plt.grid(lw=0.3)
plt.savefig("vx.pdf")
plt.show()
```

```python
fig2 = plt.figure(figsize=(8, 6), dpi=400)

plt.plot(tAn, vyAn, label="$v_y$ Analytical", ls = "-")
plt.plot(tRK, vyRK, label="$v_y$ RungeKutta", ls = "-.")
plt.xlabel("Time [s]")
plt.ylabel("Velocity [m/s]")
plt.legend()
plt.grid(lw=0.3)
plt.savefig("vy.pdf")
plt.show()
```
